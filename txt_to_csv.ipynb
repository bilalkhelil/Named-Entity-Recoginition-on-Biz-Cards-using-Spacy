```python
import pandas as pd
```

```python
#delting empty lines and adding a dot at the end of each txt file (only execute once)

for i in range(1,64):
    index = '00' + str(i)
    index = index[-3:]
    with open('extracted_text/train/'+index+'.txt', 'r') as infile:
        data = infile.read()
    infile.close()
    with open('extracted_text/train/'+index+'.txt', 'w') as infile:
        infile.write(data[:-2])
    with open("extracted_text/train/"+index+'.txt', 'a') as infile:
        infile.writelines('\n.\n')
        infile.close()
```

```python
#remove accents (requires original files to be utf-8 encoding)

for i in range(1,64):
    print('index is', i)
    index = '00' + str(i)
    index = index[-3:]
    with open('extracted_text/train/'+index+'.txt', 'r', encoding='utf-8') as file:
        data = file.read()
        print(data)
        accent = ['é', 'è', 'ê', 'à', 'ù', 'û', 'ç', 'ô', 'î', 'ï', 'â', 'É', 'È', 'Ê', 'À', 'Ù', 'Û', 'Ç', 'Ô', 'Î', 'Ï', 'Â']
        without_accent = ['e', 'e', 'e', 'a', 'u', 'u', 'c', 'o', 'i', 'i', 'a', 'E', 'E', 'E', 'A', 'U', 'U', 'C', 'O', 'I', 'I', 'A']
        i = 0
        while i < len(accent):
            data = data.replace(accent[i], without_accent[i])
            i += 1
        print(data)
    with open('extracted_text/train/'+index+'.txt', 'w', encoding='utf-8') as file:
        file.write(data)
```

```python
#concatenate individual files

with open('raw_train_data.txt', 'w') as outfile:
    for i in range(1,64):
        index = '00' + str(i)
        index = index[-3:]
        with open("extracted_text/train/"+index+'.txt') as infile:
            for line in infile:
                outfile.write(line)
outfile.close()
```

```python
#splitting each line into individual words

f = open('raw_test_data.txt', 'r')
f2 = open('raw_test_data_split.txt', 'w')
words = f.read().split()
words
for word in words:
    f2.write(word+'\n')
f.close()
f2.close()
```

```python
#txt file to DataFrame

f3 = open('raw_test_data_split.txt', 'r')
wordslist = f3.read().split()
Os = []
for i in range(len(wordslist)):
    Os.append('O')
df = pd.DataFrame({'Word':wordslist, 'Tag':Os})
```

```python
#saving to csv
df.to_csv('test_dataset_BIO.csv', index=False)
```
