```python
import spacy
from sklearn.metrics import classification_report
import pandas as pd
```

```python
nlp = spacy.load('results')
```

```python
pred_ents = []

for i in range(64, 79):
    pos = []
    dico = {}
    index = '00' + str(i)
    index = index[-3:]
    with open('extracted_text/test/'+index+'.txt', 'r', encoding = 'utf-8-sig') as file:
        BCtext = file.read()
        BCtext_line = BCtext.replace('\n', ' ')
        result = nlp(BCtext_line)
        for entity in result.ents:
            dico[entity.start_char] = entity.label_
        for n in range(len(BCtext_line)):
            if BCtext_line[n] == ' ':
                pos.append(n+1)
        pos = [0] + pos[:-1]
        for i in pos:
            if i in dico.keys():
                pred_ents.append(dico[i])
            else:
                pred_ents.append('O')
```

```python
df_true_ents = pd.read_csv('test/test_dataset_BIO.csv')
true_ents = df_true_ents['Tag'].tolist()
```

```python
print(classification_report(true_ents, pred_ents))
```
